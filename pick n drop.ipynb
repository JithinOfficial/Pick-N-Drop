Libraries Used

```python
from scipy.cluster.vq import kmeans,whiten,vq
import numpy as np
import sklearn
from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd
```

Reading datasets

```python
a=pd.read_csv(r"H:\my works pros\projects\abcd cardata.csv")
b=pd.read_csv(r"H:\my works pros\projects\myproject tr.csv")
b
```

create 'first' list to store latitude,longitudde details 

```python

first=b[['Latitude','Longitude']]
#first
```

Created odf flag to know whether the rider is  already in a ride or not & c is assigned with first list values

```python
odf=np.array(a['on duty flag'])
c=np.array(first)
```

# Label Encoding

encodeing 'street' into int values to fit

```python

en=preprocessing.LabelEncoder()
b['Street']=en.fit_transform(b['Street'])
seconds=np.array(b['Street'])
a['Street']=en.fit_transform(a['Street'])
seconda=np.array(a['Street'])
seconda
```

# K-means clustering

clustering 'streets' using k-means clustering

```python

centroids,v=kmeans(c,8)
classi,vc=vq(c,centroids)
classi
```

Created Rider-array to store assigned driver details


```python
rider_array=np.array(a[['Make','Model','Mileage (kmpl)','Seating capacity','Car class (personal opinion)','Registration number']])
rider_array
```

K-means Cluster is added to the datasets as 'Clouster Code'

```python
a['Clouster code']=classi
a
b['Clouster code']=classi
b
```

# finding nearest neighbour using knn

created secondary fit variable 'secondry' to store cluster-centroid values

```python
secondry=np.array(b['Clouster code'])
```

cross-validation first vs secondry

```python
x_train,x_test,z_train,z_test=cross_validation.train_test_split(first,secondry,test_size=0.2)
z_test
```

fitting(classification)

```python
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,z_train)
```

finding max accuracy

```python
acry=clf.score(x_test,z_test)
acry
```
